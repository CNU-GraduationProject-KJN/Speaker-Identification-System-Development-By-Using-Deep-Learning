```python
import hashlib;
def getMD5(text):	
    binary = text.encode('utf-8');	
    enc = hashlib.md5();	
    enc.update(binary);	
    ret = enc.hexdigest();	
    return ret
```

```python
import librosa
import os
import shutil
```

```python
def make_dir(name):
    if not os.path.isdir(name):
        os.makedirs(name)
        os.makedirs(name+"/origin")
        os.makedirs(name+"/preprocessed")
        os.makedirs(name+"/preprocessed/wav/")
        os.makedirs(name+"/preprocessed/pickle/train")
        os.makedirs(name+"/preprocessed/pickle/test")
        
    else:
        print("해당 폴더가 이미 존재합니다.")
    !ls
```

```python
origin_path = "../Audio_For_Speaker-Identification-System-Development-By-Using-Deep-Learning/test_train_data/"
preprocessed_path = "../Audio_For_Speaker-Identification-System-Development-By-Using-Deep-Learning/augmentation_train_data/"
save_path = "./train_data"

user_list = os.listdir(origin_path)
user_list.sort()
```

```python
for i, user in zip(range(len(user_list)), user_list):
    origin_file_list = os.listdir(origin_path + str(user+"/"))
    origin_file_list.sort()
    
    text =  save_path + "/" + getMD5(user)
    make_dir(text)
    
    for j, item in zip(range(len(origin_file_list)), origin_file_list):
        shutil.copy(origin_path+str(user+"/")+item, text+"/origin/"+item)
        
    preprocessed_file_list = os.listdir(preprocessed_path + str(user+"/"))
    preprocessed_file_list.sort()
    
    for j, item in zip(range(len(preprocessed_file_list)), preprocessed_file_list):
        shutil.copy(preprocessed_path+str(user+"/")+item, text+"/preprocessed/wav/"+item)
    
    
    
```
