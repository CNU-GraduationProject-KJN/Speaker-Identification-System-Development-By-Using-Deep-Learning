```python
file_path = "../vad_train_audio/"
save_path = "../augmentation3_train_data/"

user_list = os.listdir(file_path)
user_list.sort()

time_arr = []
time_arr2 = []

for i, user in zip(range(len(user_list)), user_list):
    file_list = os.listdir(file_path + str(user+"/"))
    file_list.sort()

    for j, item in zip(range(len(file_list)), file_list):

        sel_file_name = save_path + user +"/"+ item
        
        data, sr = librosa.load(sel_file_name, sr=8000, mono=True, duration=50)
        
        time = len(data)/sr
        time_arr.append(time)
        time_arr2.append(time)
        
        if time < am:
            if time < a25:
                aug = naa.SpeedAug(zone=(0,1), factor=(0.73,0.74)) 
            else:    
                aug = naa.SpeedAug(zone=(0,1), factor=(0.67,0.69))          
        else: 
            if time < a75: 
                aug = naa.SpeedAug(zone=(0,1), factor=(1.34,1.4))
            else:
                aug = naa.SpeedAug(zone=(0,1), factor=(1.48,1.54))
        
        augmented_data = aug.augment(data)
        time_arr2.append(len(augmented_data)/sr)
       
        files = os.listdir(save_path + user)
        result_name = str(len(files)+1)
        
        librosa.output.write_wav(save_path + user + "/" + result_name+".wav", augmented_data, sr)    
        print(" Saved File Name : "+ user +"_"+str(len(files)+1))
        
            
```

```python
def median(v):
    n = len(v)
    sorted_v = sorted(v)
    midpoint = n // 2

    if n % 2 == 1:
        # 데이터 포인트의 개수가 홀수면 중앙값을 반환
        return sorted_v[midpoint]
    else:
        # 데이터 포인트의 개수가 짝수면 두 중앙값의 평균을 반환
        lo = midpoint - 1
        hi = midpoint
        return (sorted(v)[lo] + sorted(v)[hi]) / 2
```
